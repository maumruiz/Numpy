# Line charts with matplotlib pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.linspace(0, 10, 10)
y = np.sin(x)

plt.plot(x,y)

plt.xlabel("Time")
plt.ylabel("Some function of time")
plt.title("My cool chart")

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x,y)

plt.xlabel("Time")
plt.ylabel("Some function of time")
plt.title("My cool chart")

plt.show()
```

# Scatterplot

```python
A = pd.read_csv('data_1d.csv', header=None)
A.head()
```

```python
A = A.as_matrix()

x = A[:,0]
y = A[:,1]

plt.scatter(x, y)
plt.show()
```

```python
x_line = np.linspace(0,100,100)
y_line = 2*x_line + 1

plt.scatter(x,y)
plt.plot(x_line, y_line)

plt.show()
```

# Histogram

```python
plt.hist(x)
plt.show()
```

```python
R = np.random.random(10000)
plt.hist(R)
plt.show()
```

```python
plt.hist(R, bins=20)
plt.show()
```

```python
y_actual = 2*x + 1
residuals = y - y_actual
plt.hist(residuals)
plt.show()
```

# Plotting images

```python
df = pd.read_csv("large_files/train.csv")
df.shape
```

We retrieve the first digit image, getting the first row and selecting every column except the column 0 because we know those aren't pixels, so we get a vector of 784 pixels

```python
M = df.as_matrix()

im = M[0, 1:]
im.shape
```

We convert this to images of 28x28 pixels

```python
im = im.reshape(28, 28)
im.shape
```

Then use the imshow method to plot an image

```python
plt.imshow(im)
plt.show()
```

And we check the label for that image

```python
M[0, 0]
```

To display the image in black and white we change the color map to gray

```python
plt.imshow(im, cmap='gray')
plt.show()
```

```python
plt.imshow(255 - im, cmap='gray')
plt.show()
```
